```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Prml/train")
(df)
```

```python
df.isnull().sum().sum()
```

```python
df_x = df.drop(columns="class")
df_x
```

```python
df_y = df["class"]
df_y
```

```python
import lightgbm as gbm

mod = gbm.LGBMClassifier(learning_rate=0.35)
mod.fit(df_x, df_y)
```

```python
test_d = pd.read_csv("/content/drive/MyDrive/Prml/test")
test_d
```

```python
pred = mod.predict_proba(test_d)
pred
```
