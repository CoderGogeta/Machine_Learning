```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
head = [i for i in range(1,58)]
head.append('class')
```

```python
data = pd.read_csv('/content/drive/MyDrive/spam.data', names = head)
data
```

```python
data.isna().sum().sum()
```

```python
from sklearn.model_selection import train_test_split as tts
x = data.drop[columns=['class']]
y = data[['class']]
x
```

```python
x_train,x_test,y_train,y_test = tts(x, y, test_size=0.3)
x_train
```

#a) Linear kernel

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model_lin = SVC(kernel='linear')
```

```python
from sklearn.svm import LinearSVC
```

```python
l=[1,0.9,0.8,0.7,0.6,0.5,0.4,0.3,0.2,0.1]
```

```python
for i in l:
  model = LinearSVC(C=i)
  model.fit(x_train, y_train)
  prd = model.predict(x_test)
  print(accuracy_score(y_test, prd))
model_lin.fit(x_train, y_train)
```

```python
pred_lin = model_lin.predict(x_test)
pred_lin
accuracy_score(y_test, pred_lin)
```

```python
for i in range(1,11):
  model_lin = SVC(kernel='linear', C=i)
  model_lin.fit(x_train, y_train)
  pred_lin = model_lin.predict(x_test)
  pred_lin
  print(accuracy_score(y_test, pred_lin))
```

```python
li=[1,0.9,0.8,0.7,0.6,0.5,0.4,0.3,0.2,0.1]
for i in li:
  model_lin = SVC(kernel='linear', C=i)
  model_lin.fit(x_train, y_train)
  pred_lin = model_lin.predict(x_test)
  pred_lin
  print(accuracy_score(y_test, pred_lin))
```

#b) Quadratic kernel

```python
model_qd = SVC(kernel='poly', degree=2)
# For quadratic kernel I have taken polynomial kernel with degree 2.
model_qd.fit(x_train, y_train)
```

```python
pred_qd = model_qd.predict(x_test)
pred_qd
accuracy_score(y_test, pred_qd)
```

```python
for i in range(1,101):
  model_qd = SVC(kernel='poly', degree=2, C=i)
  model_qd.fit(x_train, y_train)
  pred_qd = model_qd.predict(x_test)
  pred_qd
  print(accuracy_score(y_test, pred_qd))
```

#c) RBF kernel

```python
model_rbf = SVC(kernel='rbf')
model_rbf.fit(x_train, y_train)
```

```python
pred_rbf = model_rbf.predict(x_test)
pred_rbf
accuracy_score(y_test, pred_rbf)
```

```python
model_rbf = SVC(kernel='rbf', C=5)
model_rbf.fit(x_train, y_train)
```

```python
for i in range(1,1001):
  model_rbf = SVC(kernel='rbf', C=i)
  model_rbf.fit(x_train, y_train)
  pred_rbf = model_rbf.predict(x_test)
  pred_rbf
  print(accuracy_score(y_test, pred_rbf))
```
